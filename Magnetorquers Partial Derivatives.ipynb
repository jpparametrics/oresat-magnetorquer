To begin finding the partial derivatives of $N$, we will start by breaking up $N$ into two functions. This will make calculating the partial derivatives via the quotient rule more straight forward.

$$
N(l,r)
=
\frac{f(r,l)}{g(r,l)}
$$

$$
\begin{align}
    f(l,r)
    & =
    4
    \left[

          \frac{l}{r}
      \right)
      -1
    \right]
    \\
    & =
    4
    \ln\left(
        \frac{l}{r}
    \right)
    -4
    \\
    & = 
    4 \ln l -4 \ln r - 4
\\\\
    g(l,r)
    & =
    \left(
        \frac{l}{r}
    \right)
    ^2-4
    \ln\left(
        \frac{l}{r}
    \right)
    \\
    & =
    \frac{l^2}{r^2}
    - 4 \ln l + 4 \ln r 
\end{align}
$$

We will now also calculate the partial derivatives of $f(l,r)$ and $g(l,r)$, as well as $(g(l,r))^2$
